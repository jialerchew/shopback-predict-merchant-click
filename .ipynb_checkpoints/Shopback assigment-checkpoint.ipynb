# Preparing data

### Import data and set data types
Load data and change the default data type to either string or datetime depending on the nature of the columns.

```python
clicks = pd.read_parquet('data/clicks.parquet', engine='pyarrow')
stores = pd.read_parquet('data/stores.parquet', engine='pyarrow')
users = pd.read_parquet('data/users.parquet', engine='pyarrow')
```

```python
clicks = clicks.astype({
    'id':str,
    'user_id':str,
    'store_id':str,
    'device':str,  
    'platform':str,
    'channel':str,
    'created_at':'datetime64[ns]',
})

stores = stores.astype({
    'id':str,
    'merchant_id':str,
    'start_at':'datetime64[ns]',
    'end_at':'datetime64[ns]',  
    'display_text':str,
    'is_searchable':str
})

users = users.astype({
    'id':str,
    'signup_datetime':'datetime64[ns]',
    'lifetime_first_merchant_id':str,
    'lifetime_first_purchase_datetime':'datetime64[ns]',  
    'account_referral':str,
})
```

### Join tables

```python
# rename columns for better clarity - making the column names as verbatim as possible
stores = stores.rename(columns={
    'id':'store_id',
    'merchant_id':'store_merchant_id',
    'start_at':'store_start_at',
    'end_at':'store_end_at',
    'display_text':'store_display_text',
    'is_searchable':'store_is_searchable',
    'expected_life':'store_expected_life'
})

users = users.rename(columns={
    'signup_datetime':'user_signup_datetime',
    'lifetime_first_merchant_id':'user_lifetime_first_merchant_id',
    'lifetime_first_purchase_datetime':'user_lifetime_first_purchase_datetime',
    'account_referral':'user_account_referral',
    'user_first_purchase_since_signup':'first_purchase_since_signup'
})
users = users.rename(columns={'id':'user_id'})
```

```python
# derive cashback percentage and value based on display_text
stores['cashback_percentage'] = stores['store_display_text'].apply(lambda x: re.search(r'(\d+(\.\d+)?)(?:%)|$', x).group(1)).fillna(0).astype(float)
stores['cashback_value'] = stores['store_display_text'].apply(lambda x: re.search(r'(\d+(\.\d+)?)(?!%)(?=\s)|$', x).group(1)).fillna(0).astype(float)
```

```python
stores['cashback_value'].value_counts()
```

There are only 2 stores withh numerical cashback values, can ignore the new cashback_value column

```python
# perform left join to clicks table as per relationship diagram above
clicks = clicks.merge(stores, on='store_id', how='left')
clicks = clicks.merge(users, on='user_id', how='left')
```

```python
# derive column to map categorize channels to 10 top channels for analysis (50+ values are unconstructive for visualization)
top_channels=[
    'paid_ins_organic',
    'Direct',
    'Extension',
    'Paid Search',
    'Display',
    'paid_ins_af_banner',
    'Newsletter',
    'paid_ins_naver',
    'Organic Search',
    'paid_rtg_criteonew_int'
]

clicks['channel_category'] = 'Non-categorized'
clicks.loc[clicks['channel'].isin(top_channels), 'channel_category'] = clicks['channel']
```

```python
# derive more columns for analysis and modelling
clicks['store_expected_life'] = clicks['store_end_at'] - clicks['store_start_at']
clicks['user_first_purchase_since_signup'] = clicks['user_lifetime_first_purchase_datetime'] - clicks['user_signup_datetime']
clicks['same_as_first_purchase_merchant'] = clicks['user_lifetime_first_merchant_id'] ==  clicks['store_merchant_id']
clicks['day'] = clicks['created_at'].dt.day_name()
clicks['date'] = clicks['created_at'].dt.date
clicks['year'] = clicks['created_at'].dt.year
clicks['month'] = clicks['created_at'].dt.month
clicks['hour'] = clicks['created_at'].dt.hour
clicks['time'] = clicks['created_at'].dt.time
```

### 2) What are the distribution of clicks of merchant for month March 2021?
Ans: The distribution of clicks by merchant for March 2021 is presented below in a Pareto diagram (omitted the last 10% of click click contribution for visualization clarity) and a table. Out of 537 stores, 36 stores (7%) already make up 90% on the clicks among the total.

```python
clicks_march_2021 = clicks[(clicks['month']==3) & (clicks['year']==2021)]
clicks_march_2021_by_merchant = clicks_march_2021.groupby(['store_merchant_id'], as_index=False).agg(
    clicks_sum = pd.NamedAgg(column='id', aggfunc='count')
)
clicks_march_2021_by_merchant = clicks_march_2021_by_merchant.sort_values(['clicks_sum'],ascending=False)
clicks_march_2021_by_merchant['cumpercentage'] = clicks_march_2021_by_merchant['clicks_sum'].cumsum()/clicks_march_2021_by_merchant['clicks_sum'].sum()*100
clicks_march_2021_by_merchant = clicks_march_2021_by_merchant.reset_index()

fig, ax = plt.subplots()
df_plot = clicks_march_2021_by_merchant.head(36)
ax.bar(df_plot.index, df_plot['clicks_sum'], color='C0')
ax2 = ax.twinx()
ax2.plot(df_plot.index, df_plot['cumpercentage'], color='C1', marker='D', ms=7)
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis='y', colors='C0')
ax.set_xlabel('Merchant ranking', fontsize=10)
ax.set_ylabel('Number of clicks', fontsize=10)
ax2.tick_params(axis='y', colors='C1')
ax2.set_ylabel('Cumulative percentage of clicks', fontsize=10)
    
fig.set_size_inches(20,4)
# plt.show()
```

### 4) What is a typical click profile (highs and lows) thoughout a week? At which hour and day of the week has the most clicks?
Ans: Profiles are presented in several perspectives below. 5PM - 6PM at Monday has the highest clicks. For a more accurate 'typical' profile, we could also remove outlier dates such as events and festivals that would skew the data.

It is crucial to create a timeline 'frame' for our data to fit to prevent 'time skipping' in the time axis during vizualization because there could be no data at certain hour.

```python
# create time frame of day and hour
day_list = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
hour_list = list(range(0,23))
day_hour_list = list(itertools.product(day_list, hour_list))
day_hour_frame = pd.DataFrame(day_hour_list, columns=['day','hour']) 
date_range = pd.date_range('2021-01-01', '2021-03-31', freq='D')
date_frame = pd.DataFrame({'date': date_range}) 
```

```python
clicks_by_date = clicks.groupby(['date','day','hour','channel_category','device'], as_index=False).agg(
    clicks_sum = pd.NamedAgg(column='id', aggfunc='count')
)

clicks_by_date['date'] = clicks_by_date['date'].astype('datetime64[ns]')
clicks_by_date = date_frame.merge(clicks_by_date, on=['date'], how='left')

clicks_by_date['day'] = clicks_by_date['day'].astype('category')
clicks_by_date['day'].cat.set_categories(day_list, inplace=True)
clicks_by_date = clicks_by_date.sort_values(['date','hour'],ascending=True)
```

### 6) How likely a user will go back to the same merchant? - A measure of customers' loyalty from click perspective

The loyalty measure of a user is defined as (clicks for the most frequently visited merchant of the user)/total clicks of the user. If a user has 100%, it indicates that this user clicks on the same merchant, and has not clicked ona new merchant. On the other hand, 0% means the user has never click into any previously visited merchants.

Due to the limit of data for this exercise, this metrics is might not be useful as we only have 3 months data to approximate the measure.

On average, a user  have 50-60% tendency to revisit the same merchant. This is obviously an over-generalization because we only have 3 months worth of data for this exercise. The metrics need to be fine-tuned be more useful, such as to incorporate the recency of the visits and the purpose of visit (purchase or not) etc.

```python
# aggregate by user and merchant to calculate most clicks for a merchant by user
clicks_by_user_merchant = clicks[['user_id','store_merchant_id','id']].groupby(['user_id','store_merchant_id'], as_index=False).agg(
    clicks_sum = pd.NamedAgg(column='id', aggfunc='count')
)
clicks_by_user_merchant_most_clicks = clicks_by_user_merchant.sort_values(['user_id','clicks_sum'], ascending=False).drop_duplicates('user_id')
clicks_by_user_merchant_most_clicks = clicks_by_user_merchant_most_clicks.rename(columns = {'clicks_sum':'most_clicks_in_merchant','store_merchant_id':'most_visted_merchant_id'})
```

```python
# aggregate by user to get revisitor list
clicks_by_user = clicks[['user_id','id']].groupby(['user_id'], as_index=False).agg(
    clicks_sum = pd.NamedAgg(column='id', aggfunc='count')
)
clicks_by_revisitor = clicks_by_user[clicks_by_user['clicks_sum']>1]

clicks_by_revisitor = clicks_by_revisitor.merge(clicks_by_user_merchant_most_clicks, on='user_id', how='left')
clicks_by_revisitor['loyalty'] = clicks_by_revisitor['most_clicks_in_merchant']/clicks_by_revisitor['clicks_sum']
clicks_by_revisitor = clicks_by_revisitor.sort_values('loyalty', ascending=False)
```

```python
clicks_by_revisitor
```

```python
print("Median of user's loyalty : {0:1,.2f}%".format(np.median(clicks_by_revisitor['loyalty'])*100))
print("Mean of user's loyalty : {0:1,.2f}%".format(np.mean(clicks_by_revisitor['loyalty'])*100))
```

```python
fig, ax = plt.subplots()
df_plot = clicks_by_revisitor['loyalty']
ax = df_plot.plot.hist(figsize=(10,3), color='C0',bins=20)
ax.tick_params(axis='y', colors='C0')
ax.set_xlabel('Loyalty', fontsize=10)
ax.set_ylabel('No. of user', fontsize=10)

plt.show()
```

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x,y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
```

```python
def conditional_entropy(x,y):
    # entropy of x given y
    y_counter = Counter(y)
    xy_counter = Counter(list(zip(x,y)))
    total_occurrences = sum(y_counter.values())
    entropy = 0
    for xy in xy_counter.keys():
        p_xy = xy_counter[xy] / total_occurrences
        p_y = y_counter[xy[1]] / total_occurrences
        entropy += p_xy * math.log(p_y/p_xy)
    return entropy

def theils_u(x,y):
    s_xy = conditional_entropy(x,y)
    x_counter = Counter(x)
    total_occurrences = sum(x_counter.values())
    p_x = list(map(lambda n: n/total_occurrences, x_counter.values()))
    s_x = ss.entropy(p_x)
    if s_x == 0:
        return 1
    else:
        return (s_x - s_xy) / s_x
```
